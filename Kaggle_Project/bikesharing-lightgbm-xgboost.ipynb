# **Feature Engineering**

```python
data_path='/kaggle/input/bike-sharing-demand/'
train=pd.read_csv(data_path +'train.csv')
test=pd.read_csv(data_path +'test.csv')
```

#### **weather=4 data delete** 

```python
train=train[train['weather']!=4] # 이상치 제거 
data=pd.concat([train, test], ignore_index=True)
data.tail()
```

#### **make year, hour, weekday feature** 

```python
# year, hour, weekday feature make
data['datetime']=pd.to_datetime(data['datetime'])
data['year']=data['datetime'].dt.year
data['hour']=data['datetime'].dt.hour
data['weekday']=data['datetime'].dt.weekday
data.head()
```

#### **delete datetime, windspeed, casual, registered feature**

```python
drop_list=['datetime','casual','registered']
data = data.drop(drop_list, axis=1)
data.head()
```

```python
X_train=data[~pd.isnull(data['count'])]
X_test=data[pd.isnull(data['count'])]
X_train = X_train.drop(['count'], axis=1)
X_test = X_test.drop(['count'], axis=1)
y_train=np.log(train['count'])
```

```python
X_train.head()
```

```python
from sklearn.model_selection import train_test_split
# model early stopping valid data 
X_tr, X_val, y_tr, y_val=train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

# **RMSLE score**

```python
def RMSLE(y_real, y_pred):
    L1 = np.log1p(np.exp(y_real))
    L2 = np.log1p(np.exp(y_pred))    
    calc = (L1 - L2) ** 2
    return np.sqrt(np.mean(calc))
```

```python
# RMSLE score 
from sklearn.metrics import make_scorer
rmsle_score=make_scorer(RMSLE)
```

# **LightGBM + Optuna**

```python
from lightgbm import LGBMRegressor
import optuna
from optuna.samplers import TPESampler

def objective_func(trial):
    params = {
        'num_leaves': trial.suggest_int('num_leaves',2, 30),
        'max_depth': trial.suggest_int('max_depth',2, 30),
        'learning_rate': trial.suggest_loguniform("learning_rate", 1e-3, 1e-1),
        'n_estimators': trial.suggest_int('n_estimators', 100, 800),
        'min_child_samples': trial.suggest_int('min_child_samples', 2, 30),
        'subsample': trial.suggest_discrete_uniform('subsample', 0.3, 1, 0.1)
    }
    model=LGBMRegressor(**params)
    lgbm_model = model.fit(X_tr, y_tr, eval_set=[(X_val, y_val)], verbose=False, early_stopping_rounds=25)
    score = RMSLE(y_val, lgbm_model.predict(X_val))
    return score 
```

```python
sampler = TPESampler(seed=10)
study = optuna.create_study(direction='minimize', sampler=sampler)
study.optimize(objective_func, n_trials=50)
trial = study.best_trial
trial_params = trial.params
print('Best Trial: score {},\n params {}'.format(trial.value, trial_params))
```

```python
from optuna import visualization
optuna.visualization.plot_param_importances(study)
```

# **LightGBM OOF (5-Fold)**

```python
from sklearn.model_selection import cross_val_score
lgbm_best=LGBMRegressor(**trial_params)
score=cross_val_score(lgbm_best, X_tr, y_tr, cv=5, scoring=rmsle_score, verbose=True)
print(score)
print(np.mean(score))
```

```python
lgbm_best.fit(X_train, y_train)
```

# **LightGBM Feature Importance**

```python
from lightgbm import plot_importance
f,ax=plt.subplots(1,1,figsize=(12,10))
plot_importance(lgbm_best, ax=ax, max_num_features=15);
```

# **XGBoost + Optuna** 

```python
from xgboost import XGBRegressor
import optuna
from optuna.samplers import TPESampler

def objective_func(trial):
    params = {
        'max_depth': trial.suggest_int('max_depth',2, 30),
        'learning_rate': trial.suggest_loguniform("learning_rate", 1e-3, 1e-1),
        'n_estimators': trial.suggest_int('n_estimators', 100, 800),
        'min_child_weight': trial.suggest_int('min_child_weight', 2, 30),
        'colsample_bytree': trial.suggest_discrete_uniform('colsample_bytree', 0.3, 1, 0.1)
    }
    model=XGBRegressor(**params)
    xgb_model = model.fit(X_tr, y_tr, eval_set=[(X_val, y_val)], verbose=False, early_stopping_rounds=25)
    score = RMSLE(y_val, xgb_model.predict(X_val))
    return score 
```

```python
sampler = TPESampler(seed=10)
study = optuna.create_study(direction='minimize', sampler=sampler)
study.optimize(objective_func, n_trials=50)
trial = study.best_trial
trial_params = trial.params
print('Best Trial: score {},\nparams {}'.format(trial.value, trial_params))
```

```python
from optuna import visualization
optuna.visualization.plot_param_importances(study)
```

# **XGBoost OOF (5-Fold)**

```python
from sklearn.model_selection import cross_val_score
xgb_best=XGBRegressor(**trial_params)
score=cross_val_score(xgb_best, X_tr, y_tr, cv=5, scoring=rmsle_score, verbose=True)
print(score)
print(np.mean(score))
```

```python
xgb_best.fit(X_train, y_train)
```

# **XGBoost Feature Importance**

```python
from xgboost import plot_importance
f,ax=plt.subplots(1,1,figsize=(12,10))
plot_importance(xgb_best, ax=ax, max_num_features=15);
```

# **Ensemble**

```python
y_ret=0.9*lgbm_best.predict(X_test)+0.1*xgb_best.predict(X_test)
```

# **Submit**

```python
submission['count']=np.exp(y_ret)
submission.to_csv('submission.csv', index=False)
```
